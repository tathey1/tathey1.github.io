```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
riprev = 1
rjprev = 1

ricurs = np.arange(3)
rjcurs = np.arange(3)

njs = np.arange(1,5)
nis = np.arange(1,10)

ricurs_data = []
rjcurs_data = []
nis_data = []
njs_data = []
gds_data = []

for ricur in ricurs:
    for rjcur in rjcurs:
        for nj in njs:
            for ni in nis:
                gd = (ni*riprev - nj*rjprev + ni*(nj - 1)*rjcur - nj*(ni - 1)*ricur) / (ni + nj)
                ricurs_data.append(ricur)
                rjcurs_data.append(rjcur)
                nis_data.append(ni)
                njs_data.append(nj)
                gds_data.append(gd)

data = {"ricur": ricurs_data, "rjcur": rjcurs_data, "ni": nis_data, "nj": njs_data, "Goal Differential Needed for Team i": gds_data}
df = pd.DataFrame(data=data)
```

```python
plt.rcParams.update({'font.size': 16})
fig, axes = plt.subplots(3, 3, figsize=(15, 15), sharey=True, sharex=True)

for ricur in ricurs:
    for rjcur in rjcurs:
        sns.lineplot(ax = axes[ricur, rjcur], data=df[df["ricur"] == ricur][df["rjcur"] == rjcur], y="Goal Differential Needed for Team i", x="ni", hue="nj")
        axes[ricur, rjcur].set_title(fr"Current rankings $r_i$={ricur}, $r_j$={rjcur}")
```
